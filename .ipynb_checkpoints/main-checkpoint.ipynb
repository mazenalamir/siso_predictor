## Generate the data by simulating the dynamics

```python
import numpy as np
from generate_data import generate_random_excitation, simulate_OL

t, U = generate_random_excitation()
x0 = np.array([0.2, 0.1, 0.1])
X = simulate_OL(x0, U)
y = X[:,1]
```

## Use the model to build predictor

```python
from siso_predictor import learn_model

def ydef(xpast=None, upast=None, xplus=None, uplus=None):
    
    return xplus.mean()-xpast.mean()

sol = learn_model(y=y, 
                  U=U,
                  ydef=ydef,
                  N=100, 
                  n_clusters=3, 
                  nJump=1,  
                  max_leaf_nodes=1200,
                 test_size=0.33,
                 validation_mode='all',
                 plot=True)
```

## Test the designed predictor

```python
from siso_predictor import construct_regressor, predict

R = construct_regressor(N=sol.N, y=y, U=U, nJump=1)

ind = range(20,50)

Ypast = R.Xpast[ind,:]
Upast = R.Upast[ind,:]
Uplus = R.Uplus[ind,:]
Yplus = R.Xplus[ind,:]
z = np.array([ydef(Ypast[i,:], Upast[i,:], Yplus[i,:], Uplus[i,:]) for i in range(len(ind))])

z_predicted = predict(Ypast, Upast, Uplus, sol)
print(f"Mean_abs_error = {abs(z-z_predicted).mean()}")
```
